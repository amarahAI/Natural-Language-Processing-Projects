```python
key = "2fd4bbd11103453694109bae4c343040"
endpoint = "https://ds1textanalytics.cognitiveservices.azure.com/"


from azure.ai.textanalytics import TextAnalyticsClient
from azure.core.credentials import AzureKeyCredential

# Authenticate the client using your key and endpoint 
def authenticate_client():
    ta_credential = AzureKeyCredential(key)
    text_analytics_client = TextAnalyticsClient(
            endpoint=endpoint, 
            credential=ta_credential)
    return text_analytics_client

client = authenticate_client()

# Example function for detecting sentiment in text
non_empty_lines = [news for news in emails.split('---') if news.strip() != ""]

doc_overallscorepos=[]
doc_overallscoreneg=[]
doc_overallscoreneu=[]
sentance = []
sentence_sentiment = []
sentence_con_score_pos = []
sentence_con_score_neg = []
sentence_con_score_neu = []
doc_sentiments=[]

for i in non_empty_lines:
          
        def sentiment_analysis_example(client):
            global doc_sentiments, doc_overallscorepos,doc_overallscoreneg,doc_overallscoreneu, sentance, sentence_sentiment,sentence_con_score_pos, sentence_con_score_neg,sentence_con_score_neu
        
            documents = [i]

            response = client.analyze_sentiment(documents=documents)[0]
            print("Document Sentiment: {}".format(response.sentiment))
            doc_sentiments.append(response.sentiment)
            doc_overallscorepos.append(response.confidence_scores.positive)
            doc_overallscoreneg.append(response.confidence_scores.neutral)
            doc_overallscoreneu.append(response.confidence_scores.negative)
            print("Overall scores: positive={0:.2f}; neutral={1:.2f}; negative={2:.2f} \n".format(
                    response.confidence_scores.positive,
                    response.confidence_scores.neutral,
                    response.confidence_scores.negative,
            ))
            for idx, sentence in enumerate(response.sentences):
                print("Sentence: {}".format(sentence.text))
                sentance.append(sentence.text)
                print("Sentence {} sentiment: {}".format(idx+1, sentence.sentiment))
                sentence_sentiment.append( sentence.sentiment)

                sentence_con_score_pos.append(sentence.confidence_scores.positive)
                sentence_con_score_neg.append(sentence.confidence_scores.negative)
                sentence_con_score_neu.append(sentence.confidence_scores.neutral)
                print("Sentence score:\nPositive={0:.2f}\nNeutral={1:.2f}\nNegative={2:.2f}\n".format(
                        sentence.confidence_scores.positive,
                        sentence.confidence_scores.neutral,
                        sentence.confidence_scores.negative,
                ))
            return doc_sentiments,doc_overallscorepos,doc_overallscoreneg,doc_overallscoreneu, sentance, sentence_sentiment, sentence_con_score_pos, sentence_con_score_neg, sentence_con_score_neu


        sentiment_analysis_example(client)

        print ('----------------------------------------------------------------------------')
    


```
